```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
fruits = ['Apple','Banana','Mango','Orange','Grapes']
price = [100,30,120,50,85]
```

```python
fruits = ['Apple','Banana','Mango','Orange','Grapes']
price = [100,30,120,50,85]

plt.pie(price,labels=fruits)
plt.show()
```

```python
fruits = ['Apple','Banana','Mango','Orange','Grapes']
price = [100,30,120,50,85]

plt.pie(price,labels=fruits,autopct='%0.1f%%')
plt.show()
```

```python
fruits = ['Apple','Banana','Mango','Orange','Grapes']
price = [100,30,120,50,85]

plt.pie(price,labels=fruits,autopct='%0.1f%%',shadow=True)
plt.show()
```

```python
fruits = ['Apple','Banana','Mango','Orange','Grapes']
price = [100,30,120,50,85]

plt.pie(price,labels=fruits,autopct='%0.1f%%',explode=(0,0,0.1,0,0))
plt.show()
```

```python
fruits = ['Apple','Banana','Mango','Orange','Grapes']
price = [100,30,120,50,85]

plt.pie(price,labels=fruits,autopct='%0.1f%%',explode=(0.2,0,0.2,0,0))
plt.show()
```

```python
fruits = ['Apple','Banana','Mango','Orange','Grapes']
price = [100,30,120,50,85]

plt.pie(price,labels=fruits,autopct='%0.1f%%',shadow=True,colors=['black','blue','green','yellow','red'])
plt.show()
```
