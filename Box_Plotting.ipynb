```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [1,2,3,4,5,6,7,8,9]
y = [1,2,3,4,5,4,3,2,1]
z = [6,7,8,9,8,7,6,5,4]

data = list([x,y,z])

plt.boxplot(data)
plt.show()
```

```python
x = [1,2,3,4,5,6,7,8,9]
y = [1,2,3,4,5,4,3,2,1]
z = [6,7,8,9,8,7,6,5,4]

data = list([x,y,z])

plt.violinplot(data)
plt.title('Distribution of Data')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.show()
```

```python
x = [1,2,3,4,5,6,7,8,9]
y = [1,2,3,4,5,4,3,2,1]
z = [6,7,8,9,8,7,6,5,4]

data = list([x,y,z])

plt.violinplot(data,showmedians=True)
plt.title('Distribution of Data')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.show()
```

```python
x = [1,2,3,4,5,6,7,8,9]
y = [1,2,3,4,5,4,3,2,1]
z = [6,7,8,9,8,7,6,5,4]

data = list([x,y,z])

plt.violinplot(data,showmedians=True,showmeans=True)
plt.title('Distribution of Data')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.show()
```
