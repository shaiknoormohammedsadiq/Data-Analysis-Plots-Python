```python
# 1. Bar Graph
```

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
fruits = {'Banana':30,'Orange':50,'Mango':70,'Apple':100}
names = list(fruits.keys())
price = list(fruits.values())
names,price
```

```python
plt.bar(names,price)
plt.show()
```

```python
plt.bar(names,price)
plt.title('Market prices of Fruits')
plt.xlabel('Friuts')
plt.ylabel('Prices')
plt.show()
```

```python
plt.bar(names,price,color='yellow')
plt.title('Market prices of Fruits')
plt.xlabel('Friuts')
plt.ylabel('Prices')
plt.show()
```

```python
plt.bar(names,price,color='orange')
plt.title('Market prices of Fruits')
plt.xlabel('Friuts')
plt.ylabel('Prices')
plt.show()
```

```python
plt.barh(names,price,color='green')
plt.title('Market prices of Fruits')
plt.xlabel('Friuts')
plt.ylabel('Prices')
plt.show()
```

```python
# Scatter plot
```

```python
x = [10,20,30,40,50,60,70,80,90]
a = [1,2,3,4,5,9,8,2,1]
b = [5,8,5,9,3,4,5,2,4]
plt.scatter(x,a)
plt.scatter(x,b)
plt.show()
```

```python
x = [10,20,30,40,50,60,70,80,90]
a = [1,2,3,4,5,9,8,2,1]
b = [5,8,5,9,3,4,5,2,4]
plt.scatter(x,a)
plt.scatter(x,b)
plt.legend(['a','b'])
plt.show()
```

```python
x = [10,20,30,40,50,60,70,80,90]
a = [1,2,3,4,5,9,8,2,1]
b = [5,8,5,9,3,4,5,2,4]
plt.scatter(x,a)
plt.scatter(x,b)
plt.legend(['a','b'])
plt.title('X vs a&b')
plt.xlabel('X')
plt.ylabel('a&b')
plt.show()
```

```python
x = [10,20,30,40,50,60,70,80,90]
a = [1,2,3,4,5,9,8,2,1]
b = [5,8,5,9,3,4,5,2,4]
plt.scatter(x,a,s=200)
plt.scatter(x,b,s=200)
plt.legend(['a','b'])
plt.title('X vs a&b')
plt.xlabel('X')
plt.ylabel('a&b')
plt.show()
```

```python
x = [10,20,30,40,50,60,70,80,90]
a = [1,2,3,4,5,9,8,2,1]
b = [5,8,5,9,3,4,5,2,4]
plt.scatter(x,a,s=200,marker='3')
plt.scatter(x,b,s=200,marker='1')
plt.legend(['a','b'])
plt.title('X vs a&b')
plt.xlabel('X')
plt.ylabel('a&b')
plt.show()
```

```python
x = [10,20,30,40,50,60,70,80,90]
a = [1,2,3,4,5,9,8,2,1]
b = [5,8,5,9,3,4,5,2,4]

plt.figure(figsize=(10,5))
plt.scatter(x,a,s=200)
plt.scatter(x,b,s=200)
plt.legend(['a','b'])
plt.title('X vs a&b')
plt.xlabel('X')
plt.ylabel('a&b')
plt.show()
```
