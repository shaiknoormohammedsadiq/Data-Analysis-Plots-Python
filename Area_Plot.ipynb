```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
X = [10,20,30,40,50,60,70,80,90]
Y = [1,2,5,6,3,4,7,5,9]

plt.stackplot(X,Y)
plt.show()
```

```python
x = [10,20,30,40,50,60,70,80,90]
y = [1,2,5,6,3,4,7,5,9]

plt.stackplot(x,y)
plt.plot(x,y)
plt.show()
```

```python
x = [10,20,30,40,50,60,70,80,90]
y = [1,2,5,6,3,4,7,5,9]

plt.stackplot(x,y,colors='r')
plt.plot(x,y)
plt.show()
```

```python
x = [10,20,30,40,50,60,70,80,90]
y = [1,2,5,6,3,4,7,5,9]

plt.stackplot(x,y,colors='blue')
plt.plot(x,y,color='r')
plt.show()
```
