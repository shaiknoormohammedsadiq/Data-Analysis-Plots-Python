```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x=np.arange(1,10,0.1)
y=2*x+5
```

```python
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(1,10,0.1)
y=2*x+5
plt.plot(x,y,color='r')
plt.title('Linear Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
x=np.arange(1,10,0.1)
y=2*x+5
z=3*x+10

plt.subplot(1,2,1)
plt.plot(x,y)


plt.subplot(1,2,2)
plt.plot(x,z)
plt.show()
```
